**Parts List:**

**Enclosure and Power:**
- Plastic box (6.3 × 4.33 × 3.54 inch) – [Amazon](https://www.amazon.com/dp/B09B6QJNYD)
- Dedicated 5 V power supply – [Amazon](https://www.amazon.com/dp/B09W96X88K)
- Connector from 5 V supply to breadboard – [Amazon](https://www.amazon.com/dp/B01J1WZENK)

**Anodes:**
- Copper anode (6 × 1 × 0.04 inch) – [Amazon](https://www.amazon.com/dp/B08NGFFMLD)
- Zinc anode (6 × 1 × 0.04 inch) – [Amazon](https://www.amazon.com/dp/B07PVG5342)
- Spacers (dielectric, 1 mm) – [Amazon](https://www.amazon.com/dp/B0DNYWC9C9)
- Wire (to connect zinc and copper anodes to ADS1115) – [Amazon](https://www.amazon.com/dp/B01CQJF3BS)
- 1 M Ohn resistor (wire in parallel with anodes) - [Amazon](https://www.amazon.com/dp/B0BTP6WYH1)

**Analog-to-Digital Converter:**
- Adafruit ADS1115 (high-precision, 16-bit, STEMMA QT form factor) – [Adafruit](https://www.adafruit.com/product/1085)
- Adafruit STEMMA QT HAT – [Adafruit](https://www.adafruit.com/product/4688)
- Breadboard trim potentiometer (10 kΩ) – [Adafruit](https://www.adafruit.com/product/356)
- STEMMA QT / Qwiic JST SH 4-pin cable (100 mm long) – [Adafruit](https://www.adafruit.com/product/4210)

**Light Source:**
- UV-A LEDs (3 V; various wavelenghts: 365 nm, 375 nm, 385 nm, 405 nm, 420 nm) – [Amazon](https://www.amazon.com/dp/B0C6WQWZY1)
- Heatsink (aluminum bar) – [Amazon](https://www.amazon.com/dp/B00UJ95TOE)
- Thermal glue (to mount UV-A LEDs on heatsink) – [Amazon](https://www.amazon.com/dp/B072MSXHJD)
- IRLZ44 MOSFET (transistor to turn on LEDs programmatically) - [Amazon](https://www.amazon.com/dp/B0CBKH4XGL)
- UV-C lamp (253.7 nm wavelength) – [Amazon](https://www.amazon.com/dp/B0BVWF4TZK) *(or [alternative](https://www.amazon.com/dp/B0B5CV9PB5))*
- Household soft-white bulb (non-UV)

**Extra Tools & Safety:**
- Glue gun (to mount anodes within plastic box)
- Drill and drill bits (to create a slit in rear wall for light exposure)
- Soldering iron
- 110 V electric lamp socket (to plug in UV-C and household lamp)
- **MUST HAVE: UV protection safety goggles** – [Amazon](https://www.amazon.com/dp/B08Y5JTKMQ)
